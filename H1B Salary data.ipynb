# Archived H-1B Program Data from the Office of Foreign Labor Certification

A Labor Condition Application (LCA) is used by employers as supporting evidence for the petition for an H-1B visa. Only first issuance H-1B visas are subject to the legislated numeric limitation. <br>

The Bureau of Labor Statistics' (BLS) Occupational Employment Statistics program provides estimates used to assist in setting the wage levels in the FLC wage library.

# Convert wage to yearly wage for comparison

```python
# convert wage to yearly wage for comparison
df['wage'] = np.NaN
df.loc[df['WAGE_UNIT_OF_PAY']== u'Year','wage'] = df.loc[df['WAGE_UNIT_OF_PAY']== u'Year','WAGE_RATE_OF_PAY_FROM']
df.loc[(df['WAGE_UNIT_OF_PAY']== u'Month')&(df['WAGE_RATE_OF_PAY_FROM']<=20000), 'wage'] = df.loc[(df['WAGE_UNIT_OF_PAY']== u'Month')&(df['WAGE_RATE_OF_PAY_FROM']<=20000),'WAGE_RATE_OF_PAY_FROM']*12
df.loc[(df['WAGE_UNIT_OF_PAY']== u'Bi-Weekly') &(df['WAGE_RATE_OF_PAY_FROM']<=8000),'wage'] = df.loc[(df['WAGE_UNIT_OF_PAY']== u'Bi-Weekly') &(df['WAGE_RATE_OF_PAY_FROM']<=8000) ,'WAGE_RATE_OF_PAY_FROM']*26
df.loc[(df['WAGE_UNIT_OF_PAY']== u'Week')&(df['WAGE_RATE_OF_PAY_FROM']<=4000),'wage'] = df.loc[(df['WAGE_UNIT_OF_PAY']== u'Week')&(df['WAGE_RATE_OF_PAY_FROM']<=4000),'WAGE_RATE_OF_PAY_FROM']*52
df.loc[(df['WAGE_UNIT_OF_PAY']== u'Hour')&(df['WAGE_RATE_OF_PAY_FROM']<=200), 'wage'] = df.loc[(df['WAGE_UNIT_OF_PAY']== u'Hour')&(df['WAGE_RATE_OF_PAY_FROM']<=200),'WAGE_RATE_OF_PAY_FROM']*52*40
```

```python
# data entry error, in 2015 the following date were entered incorrectly
df.loc[ df['EMPLOYMENT_START_DATE']=='2015-15-04 00:00:00','EMPLOYMENT_START_DATE']= str('2015-04-15 00:00:00')
df.loc[ df['EMPLOYMENT_START_DATE']=='2015-15-02 00:00:00','EMPLOYMENT_START_DATE']= str('2015-02-15 00:00:00')
df.loc[ df['EMPLOYMENT_START_DATE']=='2015-15-06 00:00:00','EMPLOYMENT_START_DATE']= str('2015-06-15 00:00:00')
df.loc[ df['EMPLOYMENT_START_DATE']=='2015-19-03 00:00:00','EMPLOYMENT_START_DATE']= str('2015-03-19 00:00:00')
df.loc[ df['EMPLOYMENT_START_DATE']=='2015-15-09 00:00:00','EMPLOYMENT_START_DATE']= str('2015-09-15 00:00:00')
df.loc[ df['EMPLOYMENT_START_DATE']=='2015-09-31 00:00:00','EMPLOYMENT_START_DATE']= str('2015-09-30 00:00:00')
df.loc[ df['EMPLOYMENT_START_DATE']=='2015-14-08 00:00:00','EMPLOYMENT_START_DATE']= str('2015-08-14 00:00:00')
df.loc[ df['EMPLOYMENT_START_DATE']=='2015-17-04 00:00:00','EMPLOYMENT_START_DATE']= str('2015-04-17 00:00:00')

import datetime
df['EMPLOYMENT_START']= pd.to_datetime(df['EMPLOYMENT_START_DATE'], format="%Y-%m-%d %H:%M:%S")

# change all SOC_NAME to upper case
df['SOC_NAME'] = df['SOC_NAME'].str.upper()
```
